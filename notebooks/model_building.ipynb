Build a regression model.

```python
import pandas as pd
import statsmodels.api as sm
```

```python
Bike_Stations_Joined = pd.read_csv('Bike_Stations_Joined.csv')
```

```python
Bike_Stations_Joined
```

```python
relevant_metrics = Bike_Stations_Joined[['Bike_Usage', 'FS Restaurant Distance', 'Station Distance', 'yelp_distance','available_bikes','empty_slots']].copy()
```

```python
relevant_metrics['total slots']=relevant_metrics['empty_slots']+relevant_metrics['available_bikes']
```

```python
relevant_metrics[['total slots','FS Restaurant Distance', 'Station Distance', 'yelp_distance']].corr()
```

```python
y = relevant_metrics['total slots']
X = relevant_metrics[['FS Restaurant Distance', 'Station Distance', 'yelp_distance']]
X = sm.add_constant(X)
X.head()
```

```python
model = sm.OLS(y, X) #instantiate
results = model.fit() #fit the model (this is where OLS is actually being run)
```

Provide model output and an interpretation of the results. 

```python
print(results.summary())
```

There is no real significant correlation between the distance variables in being able to predict the number of total bikes at a nearby bikestation. 
The null hypothesis of no correlation can be accepted because all three variables have p values greater than 0.05. 
This matches up with the coeff values all being relatively low (0.001 or lower). 
Overall this is a weak model in being able to predict total bikes as the adj r-squared is essentially showing that it can't predict even 1% of values (-0.012).
