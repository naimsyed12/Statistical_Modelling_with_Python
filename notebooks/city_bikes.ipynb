# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import pandas as pd
import requests
import os
import numpy as np
```

```python
##Initial request for MIAMI bike stations (citi-bike-miami)
url = 'http://api.citybik.es/v2/networks/citi-bike-miami'
res = requests.get(url)
res.json()
```

```python
res.json()['network']['stations']
```

```python
bike_stations = res.json()['network']['stations']               ##filtered through initial API results to get a diction of each station and it's corresponding sublists
```

```python
len(bike_stations)
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
latitude = []
longitude = []
available_bikes = []
empty_slots = []
name = []

for station in bike_stations:
    latitude.append(station['latitude'])
    longitude.append(station['longitude'])
    available_bikes.append(station['free_bikes'])
    empty_slots.append(station['empty_slots'])
    name.append(station['name'])
                
#gets relevant information, but can't be used for a dataframe
#for each individual bike station dictionary, in the list of bike_stations, extract value and .append()   
```

Put your parsed results into a DataFrame.

```python
data = []


```

```python
for station in bike_stations:
    latitude = station['latitude']
    longitude = station['longitude']
    available_bikes = station['free_bikes']
    empty_slots = station['empty_slots']
    name = station['name']
    data.append([latitude,longitude,available_bikes,empty_slots,name])

## copy of previous loop, but stores all indiviual list into a combined list to be used in a dataframe
```

```python
df = pd.DataFrame(data, columns=['latitude','longitude','available_bikes','empty_slots','name'] )
```

```python
df.to_csv('Bike_Stations.csv', index=False)
```
